# Lawful Joy

- The Joy of Living Within the Lawful

`````{admonition} Reading Objectives

`````

## Background
```{figure} pict/11_ray_5_topic/background.png
---
scale: 40%
align: left
---
```

In the Fifth Topic of the Eleventh Ray, Bediüzzaman Said Nursi reflects on one of life's greatest blessings: youth. He reminds readers that youth is temporary and will inevitably give way to old age and death.

Rather than seeing youth merely as a time for pleasure, he presents it as a Divine trust to be used wisely. Drawing on the teachings of the Qur'an and the revealed scriptures, he explains that true happiness is found within the lawful (licit) bounds established by God. While unlawful pleasures may appear attractive, they often bring lasting regret and suffering, whereas lawful living leads to peace, gratitude, and eternal reward.

## Main Passage

```{figure} pict/11_ray_5_topic/main_passage.png
---
scale: 40%
align: right
---
```

[Download Main Passage](11_ray_5_topic.pdf)

As is described in A Guide For Youth, there is no doubt that youth will depart; it will change into old age and death as certainly as the summer gives its place to autumn and winter, and the day changes into evening and night.


All the revealed scriptures give the good news that if <font color="red">fleeting</font>, <font color="red">transient</font> youth is spent on good works, in <font color="red">chastity</font> and within the bounds of good <font color="red">conduct</font>, it will gain for the person <font color="red">immortal</font>youth.

If, on the other hand, youth is spent on <font color="red">vice</font>, just as murder resulting from a minute's anger leads to millions of minutes of <font color="red">imprisonment</font>, so quite apart from being called to account in the <font color="red">hereafter</font>, and the <font color="red">torments</font> of the grave, and the regrets arising from their passing, and sins, and the penalties suffered in this world, the unlawful pleasures of youth contain more pain than pleasure; every youth with sense will <font color="red">corroborate</font> this from his own experience.


For example, the pains of jealousy, separation, and <font color="red">unreciprocated</font> love transform the partial pleasure to be found in <font color="red">illicit</font> love into poisonous honey.

If you want to know how they end up in hospitals due to illnesses resulting from their misspent youth, and in prison due to their excesses, and in bars and <font color="red">dens</font> of vice and the graveyard due to the distress arising from their <font color="red">unnourished</font> hearts and spirits not performing their right functions, go and ask at the hospitals, prisons, bars, and graveyards.


More than anything, you will hear the <font color="red">weeping</font> and sighs of regret at the blows youths have received as the penalty for abusing their youth, and their excesses, and <font color="red">illicit</font> pleasures.


Foremost the Qur'an, with numerous of its verses, and all the revealed scriptures and books, give the glad <font color="red">tidings</font> that if spent within the <font color="red">bounds</font> of moderation, youth is an agreeable Divine bounty and sweet, powerful means to good works, which yields the result of shining, <font color="red">immortal</font> youth in the hereafter.


Since the reality is this, and since the bounds of the licit are sufficient for enjoyment, and since an hour of unlawful pleasure leads sometimes to a punishment of one, or ten, years' imprisonment; surely it is absolutely necessary to spend the sweet bounty of youth <font color="red">chastely</font>, on the straight path, as thanks for the bounty.


## Summary 

```{figure} pict/11_ray_5_topic/summary.png
---
scale: 30%
align: left
---
```
Bediüzzaman teaches that youth is a precious but temporary blessing. When spent in faith, purity, and good character, it becomes a source of everlasting happiness in the Hereafter.

Although forbidden pleasures may offer temporary excitement, they often result in emotional pain, physical harm, regret, and accountability before God. In contrast, the lawful things that God has permitted are sufficient for genuine enjoyment and lead to a life of gratitude, self-respect, and lasting peace.

He concludes that the best way to thank God for the gift of youth is to live within the lawful bounds He has established.

<br><br><br><br><br><br><br><br>

## Reading Comprehension Questions
```{figure} pict/rc.png
---
scale: 50%
align: right
---
```
1. Why does Bediüzzaman say that youth is temporary?
1. What do the revealed scriptures promise to those who spend their youth righteously?
1. What comparison does he make between youth and the changing seasons?
1. What consequences can result from abusing one's youth?
1. Why does he compare unlawful pleasure to poisonous honey?
1. Which places does he mention as evidence of the consequences of misused youth?
1. What does he say the Qur'an teaches about youth?
1. Why are the lawful bounds sufficient for enjoyment?
1. What can a single hour of unlawful pleasure sometimes lead to?
1. According to the passage, how should young people show gratitude for the blessing of youth?


## Critical Thinking Questions
```{figure} pict/ct.png
---
scale: 50%
align: left
---
```

1. Why do many young people feel that youth will last forever?
1. How do today's media and culture influence the way young people use their youth?
1. What does it mean to view youth as a trust from God?
1. Why are short-term pleasures often more attractive than long-term rewards?
1. How can self-discipline protect a person's future?
1. What kinds of regrets might result from poor decisions made during youth?
How can lawful enjoyment lead to a happier life?
Why is gratitude an important part of using blessings wisely?
How would your daily choices change if you viewed youth as preparation for eternity?
What practical steps can young people take to benefit from their youth today?


## Glossary
```{figure} pict/glossary.png
---
scale: 80%
align: right
---
```

| Word | Definition |
|------|------------|
| Bounty | A generous gift or blessing from God. |
| Chastity | Living with purity and self-control according to moral and religious principles. |
| Conduct | A person's behavior or actions. |
| Corroborate | To confirm or support with evidence or experience. |
| Dens | Places associated with immoral or harmful activities. |
| Fleeting | Lasting for only a short time. |
| Glad Tidings | Good news or joyful promises from God. |
| Hereafter | Life after death. |
| Illicit | Forbidden by law or religion. |
| Immortal | Living forever; never ending. |
| Imprisonment | The state of being confined in prison. |
| Licit | Allowed or permitted according to law or religion. |
| Moderation | Avoiding excess; keeping within proper limits. |
| Tidings | News or information, especially good or important news. |
| Torrents | Severe or overwhelming experiences; here, intense suffering or punishment. |
| Transient | Temporary; lasting only a short time. |
| Unnourished | Not properly cared for or developed; lacking what is needed to grow healthy. |
| Unreciprocated | Not returned or shared by another person; especially referring to love or affection. |
| Vice | Immoral or sinful behavior. |
| Weeping | Crying with tears because of sadness or pain. |
