
**The Risale-i Nur Collection**

The Risale-i Nur Collection is a Qur'anic commentary (tafsir) written by Bediüzzaman Said Nursi (1877–1960). Rather than explaining the Qur'an verse by verse, it explains its teachings through themes and subjects such as belief in God, resurrection, Divine Names, worship, gratitude, morality, and the purpose of human life.

Written during a period when materialism, secular philosophy, and modern science raised new questions about faith, the Risale-i Nur presents logical, thoughtful, and evidence-based explanations that strengthen belief. Its aim is to help readers develop verified faith (iman tahqiqi)—faith based on understanding, reflection, and conviction rather than imitation.

Bediüzzaman describes one of the spiritual methods followed throughout the collection:

"One of these is the way of impotence, poverty, compassion, and reflection, from which, with my defective understanding, I have benefited."

Throughout the collection, everyday examples, comparisons, and stories are used to explain profound spiritual truths in a clear and memorable way. Bediüzzaman also advised readers not to approach the Risale-i Nur as they would a newspaper to be read quickly and set aside. Instead, it should be read slowly, thoughtfully, and repeatedly, allowing time for reflection and discussion so that its meanings can be fully appreciated.

The Risale-i Nur addresses the intellectual and spiritual needs of the modern age by answering questions about faith with reason, observation, and the teachings of the Qur'an. It seeks to transform belief from simple acceptance into confident understanding based on knowledge and clear evidence.

**The Fourteen Main Books of the Risale-i Nur**

| Book | Description |
|------|-------------|
| **The Words (Sözler)** | Explains the foundations of Islamic belief through examples, comparisons, and logical arguments. |
| **The Letters (Mektubat)** | Answers important questions about faith, worship, and daily life through a collection of letters. |
| **The Flashes (Lem'alar)** | Presents practical lessons on sincerity, patience, gratitude, morality, and spiritual growth. |
| **The Rays (Şuâlar)** | Explores deeper subjects such as resurrection, Divine Names, and responses to modern doubts about faith. |
| **The Staff of Moses (Asâ-yı Musa)** | Provides concise and convincing proofs for the existence of God and the truths of belief. |
| **Mesnevi-i Nuriye** | A collection of short reflections that nurture the heart and strengthen faith through contemplation. |
| **Signs of Miraculousness (İşârâtü'l-İ'câz)** | A thematic commentary on the opening chapters of the Qur'an, highlighting their depth and eloquence. |
| **Reasonings (Muhâkemat)** | Explains principles for correctly understanding Islam and distinguishes sound reasoning from misconceptions. |
| **Comparisons of Belief and Unbelief (İman ve Küfür Muvazeneleri)** | Compares the results of belief and unbelief, showing how faith gives meaning, hope, and purpose to life. |
| **The Barla Addendum (Barla Lâhikası)** | Letters describing the early development of the Risale-i Nur and guidance for its students. |
| **The Kastamonu Addendum (Kastamonu Lâhikası)** | Letters emphasizing sincerity, perseverance, and strengthening faith during difficult times. |
| **The Emirdağ Addendum (Emirdağ Lâhikası)** | Advice on serving faith with wisdom, patience, and hope in changing social conditions. |
| **The History of Life (Tarihçe-i Hayat)** | The biography of Bediüzzaman Said Nursi and the historical background of the Risale-i Nur. |
| **The Seal of the Unseen Confirmation (Sikke-i Tasdik-i Gaybi)** | Presents signs and evidences confirming the truths of faith and the value of the Risale-i Nur. |
